Task - 4 Image to Pencil Sketch with Python

In this task, we need to read the image in RBG format and then convert it to a grayscale image. This will turn an image into a classic black and white photo. Then the next thing to do is invert the grayscale image also called negative image, this will be our inverted grayscale image. Inversion can be used to enhance details. Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image. Since images are just arrays, we can easily do this programmatically using the divide function from the cv2 library in Python.

References: https://thecleverprogrammer.com/2020/09/30/pencil-sketch-with-python/

# Importing the necessary libraries

```python
import cv2
from cv2_plt_imshow import cv2_plt_imshow
import matplotlib.pyplot as plt
```

# Loading And Displaying the Image 

```python
image=cv2.imread(r"C:\Users\HP\Downloads\statue_of_unity_sardar_patel-U205730969126CB--621x414@LiveMint.png")
cv2_plt_imshow(image)
```

# Converting the image to greyscale image 

```python
grey_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY) 
cv2_plt_imshow(grey_image)
```

# Inverting the image

```python
inverted_image = cv2.bitwise_not(grey_image)  
cv2_plt_imshow(inverted_image)
```

# Blurring the image

```python
image_smoothing = cv2.GaussianBlur(inverted_image, (21, 21),sigmaX=0, sigmaY=0)
cv2_plt_imshow(image_smoothing)
```

# Creating the pencil sketch using the grey scale and inverted image

```python
def pencil_sketch(x, y):

    return cv2.divide(x, 255 - y, scale=256)
```

```python
pencil_sketch=pencil_sketch(grey_image,image_smoothing)
cv2_plt_imshow(pencil_sketch)
```
